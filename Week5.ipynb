## Deep Dive into NumPy

NumPy Array Operations

```python
import numpy as np

# Creating an array
arr = np.array([10, 20, 30, 40, 50])

# Performing operations
print("Sum:", np.sum(arr))
print("Mean:", np.mean(arr))
print("Standard Deviation:", np.std(arr))

# Generating a random 3x3 matrix
random_matrix = np.random.randint(1, 100, size=(3, 3))
print("\nRandom Matrix:\n", random_matrix)
```

```python
import numpy as np

# Create a NumPy array
arr = np.array([10, 20, 30, 40, 50])
print("Original Array:", arr)

# Perform basic operations
arr_sum = np.sum(arr)
arr_mean = np.mean(arr)
arr_std = np.std(arr)

print("Sum:", arr_sum)
print("Mean:", arr_mean)
print("Standard Deviation:", arr_std)

# Generate a random NumPy array
random_arr = np.random.randint(1, 100, size=(5, 5))  # 5x5 random integers
print("\nRandom Array:\n", random_arr)

# Slicing the array (first two rows)
print("\nSliced Array (first 2 rows):\n", random_arr[:2])

# Element-wise operations
arr2 = np.array([1, 2, 3, 4, 5])
print("\nElement-wise Multiplication:", arr * arr2)

# Generate a histogram of random values
import matplotlib.pyplot as plt

rand_data = np.random.randn(1000)  # Generate 1000 random numbers
plt.hist(rand_data, bins=30, color='purple', edgecolor='black', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Data')
plt.show()
```

## Mini Project: Data Analysis with Pandas & Matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Creating a sample dataset
data = {'Year': np.arange(2015, 2025),
        'Sales': np.random.randint(1000, 5000, size=10)}

df = pd.DataFrame(data)
print("Sales Data:\n", df)

# Plot sales over the years
plt.figure(figsize=(6, 4))
plt.bar(df['Year'], df['Sales'], color='green', alpha=0.6)
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Company Sales Over the Years')
plt.grid(axis='y')
plt.show()
```
