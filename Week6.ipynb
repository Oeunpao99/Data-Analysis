```python
plt.figure(figsize=(6, 4))
n, bins, patches = plt.hist(df['age'], bins=20, color='blue', edgecolor='w')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Adding labels above each bar
for i in range(len(patches)):
    plt.text(patches[i].get_x() + patches[i].get_width() / 2, n[i], str(int(n[i])), 
             ha='center', va='bottom', fontsize=10, color='black')

plt.show()
```

## breakdown: 
- n: An array containing the count of data points in each bin (height of each bar).
- bins: An array defining the edges of each bin (e.g., age ranges).
- patches: A list of bar objects representing each bar in the histogram.

Loop places the count label above each bar. Let’s break it down:
- patches[i].get_x() + patches[i].get_width() / 2:

- patches[i].get_x() retrieves the left edge of the bar.
- patches[i].get_width() gives the width of the bar.
- Adding half the width centers the label horizontally above the bar.
- n[i]:
- This represents the height of the bar (the count of data points in that bin). The label is placed slightly above the top edge of the bar.

- str(int(n[i])):

- This converts the count to a string so it can be displayed as text.
- int(n[i]) ensures there are no decimal points in the label.
- Text Alignment:
ha='center': This horizontally centers the label above the bar.
va='bottom': This places the label just above the bar’s top edge.
- Text Style:
fontsize=10: This controls the size of the label text.
color='black': This sets the color of the text to black for better readability.
